```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv("/content/Uncleaned_employees_final_dataset.csv")
```

```python
display(data)
```

```python
print(data.dtypes)
```

```python
columns = list(data.columns)
columns
```

```python
# Collecting data on mean of null values present in each column
print(data.isnull().mean())
```

```python
# Removing duplicate rows
data_updated_1=data.drop_duplicates(subset=None, keep='first', inplace=False, ignore_index=False)
```

```python
# To fill non-numeric values in columns cotaining numeric values with NAN values
numeric_cols = data_updated_1.select_dtypes(include=['int64', 'float64']).columns
for col in numeric_cols:
    data_updated_1[col] = pd.to_numeric(data_updated_1[col], errors='coerce')
```

```python
display(data_updated_1)
```

```python
for col in numeric_cols:
   if col == 'employee_id ' or col=='no_of_trainings' or col=='age' or col=='previous_year_rating' or col=='length_of_service' or col=='KPIs_met_more_than_80' or col=='awards_won' or col=='avg_training_score' :
        # remove rows that contain missing or invalid values
        data_updated_1 = data_updated_1.dropna(subset=[col])
display(data_updated_1)
```

```python
for col in columns:
  # remove rows that contain missing or invalid values
  if col=="education":
    data_updated_1 = data_updated_1.dropna(subset=[col])
display(data_updated_1)
data_updated_1.to_csv('cleaned_dataset.csv', index=False)
```
