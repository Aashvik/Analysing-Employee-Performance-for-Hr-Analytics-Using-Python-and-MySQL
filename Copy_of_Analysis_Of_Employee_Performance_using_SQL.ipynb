```python
import pandas as pd
import sqlite3
```

```python
#Create Table employee
db=sqlite3.connect("EmployeesPerfs.db")
db.execute("drop table if exists employee")
#Read and execute the SQL Statements in employee.sql file
with open('employee.sql', 'r') as sql_file:
    sql_script = sql_file.read()
try:
  db.executescript(sql_script)
  db.commit()
except:
  print("Table employee already exists")
```

```python
rows=db.execute("SELECT * FROM employee")
for i in rows:
  print(i)
```

```python
for i in db.execute("SELECT COUNT(*) FROM employee"):
   print("row count = ", i)
```

```python
#Find the average age of employees in each department and gender group. ( Round average age up to two decimal places if needed).
r1=db.execute("SELECT department, gender, ROUND(AVG(age),2) AS average_age_of_employees FROM employee GROUP BY department, gender")
for i in r1:
  print(i)
```

```python
#List the top 3 departments with the highest average training scores. ( Round average scores up to two decimal places if needed)
r2=db.execute("SELECT department, ROUND(AVG(avg_training_score),2) AS average_training_score_of_a_department FROM employee GROUP BY department ORDER BY avg_training_score DESC LIMIT 3")
for i in r2:
  print(i)
```

```python
#Find the percentage of employees who have won awards in each region. (Round percentages up to two decimal places if needed)
r3=db.execute("SELECT region, ROUND((COUNT(CASE WHEN awards_won=1 then 1 END)/(SELECT COUNT(*) FROM employee WHERE awards_won>0))*100,2) AS award_percentage FROM employee GROUP BY region HAVING award_percentage<>0;")
'''with open('task3.sql', 'r') as t3_file:
    sql_script_1 = t3_file.read()
r3_1=db.executescript(sql_script_1)'''
for i in r3:
  print(i)
```

```python
#Show the number of employees who have met more than 80% of KPIs for each recruitment channel and education level.
r4=db.execute("SELECT recruitment_channel, education,COUNT(employee_id) AS number_of_employee FROM employee WHERE KPIs_met_more_than_80>0 GROUP BY recruitment_channel, education;")
for i in r4:
  print(i)
```

```python
#Find the average length of service for employees in each department, considering only employees with previous year ratings greater than or equal to 4. ( Round average length up to two decimal places if needed)
r5=db.execute("SELECT department, ROUND(AVG(length_of_service),2) AS AVG_Length_OF_SERVICE FROM employee WHERE previous_year_rating>=4 GROUP BY department;")
for i in r5:
  print(i)
```

```python
#List the top 5 regions with the highest average previous year ratings. ( Round average ratings up to two decimal places if needed)
r6=db.execute("SELECT region, ROUND(AVG(previous_year_rating),2) AS average_of_prev_year_rating FROM employee GROUP BY region ORDER BY ROUND(AVG(previous_year_rating),2) DESC LIMIT 5;")
for i in r6:
  print(i)
```

```python
#List the departments with more than 100 employees having a length of service greater than 5 years.
r7=db.execute("SELECT department, COUNT(employee_id) AS employee_count FROM employee WHERE length_of_service>5 GROUP BY department HAVING COUNT(employee_id)>100;")
for i in r7:
  print(i)
```

```python
#Show the average length of service for employees who have attended more than 3 training, grouped by department and gender. ( Round average length up to two decimal places if needed)
r8=db.execute("SELECT department, gender, ROUND(AVG(length_of_service),2) AS average_length_of_service FROM employee WHERE no_of_trainings>3 GROUP BY department, gender;")
for i in r8:
  print(i)
```

```python
#Find the percentage of female employees who have won awards, per department. Also show the number of female employees who won awards and total female employees. ( Round percentage up to two decimal places if needed)
r9=db.execute("SELECT department, ROUND(SUM(CASE WHEN gender='f' AND awards_won>0 THEN 1 ELSE 0 END)/SUM(CASE WHEN gender='f' THEN 1 ELSE 0 END)*100,2) AS percentage_female_award_winners, SUM(CASE WHEN gender='f' AND awards_won>0 THEN 1 ELSE 0 END) AS female_awards_won, SUM(CASE WHEN gender='f' THEN 1 ELSE 0 END) AS total_females FROM employee GROUP BY department;")
for i in r9:
  print(i)
```

```python
#Calculate the percentage of employees per department who have a length of service between 5 and 10 years. ( Round percentage up to two decimal places if needed)
r10=db.execute("SELECT department,ROUND(SUM(CASE WHEN length_of_service>=5 AND length_of_service<=10 THEN 1 ELSE 0 END)/COUNT(*)*100,2) AS avg_length_of_service FROM employee GROUP BY department;")
for i in r10:
  print(i)
```

```python
#Find the top 3 regions with the highest number of employees who have met more than 80% of their KPIs and received at least one award, grouped by department and region.
r11=db.execute("SELECT department,region, COUNT(employee_id) AS total_employees FROM employee WHERE KPIs_met_more_than_80>0 AND awards_won>0 GROUP By department, region ORDER BY COUNT(employee_id) DESC LIMIT 3;")
for i in r11:
  print(i)
```

```python
#Calculate the average length of service for employees per education level and gender, considering only those employees who have completed more than 2 trainings and have an average training score greater than 75 ( Round average length up to two decimal places if needed)
r12=db.execute("SELECT education, gender, ROUND(AVG(length_of_service),2) AS average_length_of_service FROM employee WHERE no_of_trainings>2 AND avg_training_score>75 GROUP BY education, gender;")
for i in r12:
  print(i)
```

```python
#For each department and recruitment channel, find the total number of employees who have met more than 80% of their KPIs, have a previous_year_rating of 5, and have a length of service greater than 10 years.
r13=db.execute("SELECT department, recruitment_channel, COUNT(employee_id) AS total_employees FROM employee WHERE KPIs_met_more_than_80>0 AND previous_year_rating=5 AND length_of_service>10 GROUP BY department, recruitment_channel;")
for i in r13:
  print(i)
```

```python
#Calculate the percentage of employees in each department who have received awards, have a previous_year_rating of 4 or 5, and an average training score above 70, grouped by department and gender ( Round percentage up to two decimal places if needed).
r14=db.execute("SELECT department, gender, ROUND(SUM(CASE WHEN awards_won>0 AND (previous_year_rating=4 OR previous_year_rating=5) AND avg_training_score>70 THEN 1 ELSE 0 END)/COUNT(*)*100,2) FROM employee GROUP BY department, gender;")
for i in r14:
  print(i)
```

```python
#List the top 5 recruitment channels with the highest average length of service for employees who have met more than 80% of their KPIs, have a previous_year_rating of 5, and an age between 25 and 45 years, grouped by department and recruitment channel. ( Round average length up to two decimal places if needed).
r15=db.execute("SELECT department, recruitment_channel, ROUND(AVG(length_of_service),2) AS average_length_of_service FROM employee WHERE KPIs_met_more_than_80>0 AND previous_year_rating=5 AND (age>=25 AND age<=45) GROUP BY department, recruitment_channel ORDER BY average_length_of_service DESC LIMIT 5;")
for i in r15:
  print(i)
```
